```python
from pydub import AudioSegment
import os
name = "faiszal"
input_file = f"{name}.m4a"   # bisa juga mp4
output_dir = f"{name}/"
chunk_length_ms = 5000  # 5 detik

os.makedirs(output_dir, exist_ok=True)

# load audio
audio = AudioSegment.from_file(input_file)

duration = len(audio)

count = 0
for start in range(0, duration, chunk_length_ms):
    chunk = audio[start:start + chunk_length_ms]

    output_file = os.path.join(output_dir, f"{name}_{count}.wav")
    chunk.export(output_file, format="wav")

    count += 1

print("Selesai, total chunk:", count)
```
